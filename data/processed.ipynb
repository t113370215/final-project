```python
# ==============================================================================
# 【中文說明】
# 讀取原始的 YRBS_2007.csv 檔案，進行以下資料清理：
# 1. 將性別欄位（WhatIsYourSex）的 1.0 轉為 Male，2.0 轉為 Female。
# 2. 自動剔除掉性別或體重欄位中，填寫不完整或有空格（NaN）的無效資料。
# 3. 在專案目錄下自動創建 "data/processed" 兩層資料夾。
# 4. 將篩選乾淨的數據導出，命名為 "YRBS_2007_processed.csv" 存入該子資料夾中。
#
# 【English Explanation】
# Reads the original YRBS_2007.csv file and performs the following data cleaning steps:
# 1. Maps the gender column (WhatIsYourSex) codes: 1.0 to "Male" and 2.0 to "Female".
# 2. Automatically removes invalid records with missing values (NaN) in sex or weight columns.
# 3. Creates nested directories "data/processed" in the project root path.
# 4. Exports the cleaned dataset as "YRBS_2007_processed.csv" inside the data/processed folder.
# ==============================================================================

import os
import pandas as pd

print("==================================================")
print("🗂️ 資料篩選與自動導出系統 (儲存至 data/processed 資料夾)")
print("==================================================")

# 1. 讀取原始資料集
input_filename = "YRBS_2007.csv"
if not os.path.exists(input_filename):
    raise FileNotFoundError(
        f"❌ 在目前的資料夾中找不到 {input_filename}，請確認檔案放置路徑是否正確！"
    )

df_raw = pd.read_csv(input_filename)
df_processed = df_raw.copy()

weight_col = "HowMuchDoYouWeighWithoutShoesInKG"
sex_col = "WhatIsYourSex"

# 2. 轉換性別標籤：將數字 1.0 與 2.0 替換為英文名稱
df_processed[sex_col] = pd.to_numeric(df_processed[sex_col], errors="coerce")
gender_mapping = {1.0: "Male", 2.0: "Female"}
df_processed[sex_col] = df_processed[sex_col].map(gender_mapping)

# 3. 體重欄位強制轉為數值型態
df_processed[weight_col] = pd.to_numeric(
    df_processed[weight_col], errors="coerce"
)

# 4. 篩選並剔除掉「性別」或「體重」有漏填(NaN)的橫列
initial_count = len(df_processed)
df_processed = df_processed.dropna(subset=[weight_col, sex_col])
final_count = len(df_processed)
deleted_count = initial_count - final_count

print(f"📋 原始總資料筆數：{initial_count} 筆")
print(f"🧹 已自動剔除漏填/無效資料：{deleted_count} 筆")
print(f"✨ 篩選後的有效男女生總筆數：{final_count} 筆")

# ==============================================================================
# 🎯 【核心修正：自動創建兩層資料夾 data/processed 並將檔案存入】
# ==============================================================================
output_dir = os.path.join("data", "processed")
output_filename = "YRBS_2007_processed.csv"
output_path = os.path.join(output_dir, output_filename)

# os.makedirs 加上 exist_ok=True 可以一次蓋好兩層資料夾（data 和 processed）
os.makedirs(output_dir, exist_ok=True)

# 導出成全新的 CSV 檔案
df_processed.to_csv(output_path, index=False)

print("-" * 50)
print(f"✅ 檔案已成功處理完成！")
print(f"📁 儲存路徑：{output_path}")
print("==================================================")
```
